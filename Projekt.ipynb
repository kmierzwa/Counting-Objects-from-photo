```python

def g_blur(img4):
	img = img4
	for x in range(1, img.shape[0] - 1):
		for y in range(1, img.shape[1] - 1):
			mean = 0
			for i in range(x-1, x + 1):
				for j in range(y-1, y + 1):
					mean += img[i][j]
			mean /= 9
			img[x][y] = mean
	return img


def threshold(img, thresh):
	for x in range(img.shape[0]):
		for y in range(img.shape[1]):
			if img[x][y] > thresh:
				img[x][y] = 255
			else:
				img[x][y] = 0
	return img


```

```python
seeds = io.imread('ziarna.png')
seeds1 = io.imread('ziarna1.png')
seeds2 = io.imread('ziarna2.png')
seeds3 = io.imread('ziarna3.png')

b_seeds1 = color.rgb2gray(seeds1)
b_seeds1 = misc.bytescale(b_seeds1)
b_seeds1 = g_blur(b_seeds1)
b_seeds1 = threshold(b_seeds1, 35)

b_seeds1 = cv2.distanceTransform(b_seeds1,cv2.DIST_L2,0)

b_seeds1 = threshold(b_seeds1, 4)

number_of_seeds1 = measure.label(b_seeds1)
print("Znaleziono na obrazku nr 1:", number_of_seeds1.max(), "obiektów")

plt.imshow(b_seeds1, cmap='gray')
plt.show()

```
